'tengo que analizar todas las lineas "terminada", y la anterior, coger la casa y sumar el beneficio'


```python
lineas=contenido.split("\n")
def beneficio(texto):
    return float(texto.split('\t')[3].replace(",","."))*-1
def casa(texto):
    return texto.split('\t')[2]
def fecha(texto):
    return texto.split('\t')[1]
```

```python
contador=-1
for linea in lineas:
    contador +=1
    if ("terminada" in linea):
        #print(lineas[contador],"\n",lineas[contador-1])
        #extrae cantidad de beneficio
        #extrae casa
        print(fecha(lineas[contador]),casa(lineas[contador-1]),beneficio(lineas[contador]))
        #salida+="'"+fecha(lineas[contador])+"'"
        #salida+="'"+casa(lineas[contador-1])+"'"
        #salida+="'"+str(beneficio(lineas[contador]))+"'"
        #salida +=fecha(lineas[contador]+","+casa(lineas[contador-1])+','+str(beneficio(lineas[contador]))+'|'
```
